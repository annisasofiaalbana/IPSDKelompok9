```python
#Elementwise Function menggunakan kali, bagi dan tambah

import numpy as np

suhu_celsius = np.array([22, 24, 28, 19, 25, 30, 23])


suhu_fahrenheit = (suhu_celsius * 9/5) + 32

print("Rumus: (C * 9/5) + 32 diterapkan secara elementwise")
print(f"Suhu Fahrenheit (F): {suhu_fahrenheit}")
```
