Karakteristik Data

```python
import numpy as np

data = np.array([[1, 2], [3, 4], [5, 6]])

print("Isi array:\n", data)
print("Jumlah dimensi:", data.ndim)      # 2 (karena 2D)
print("Bentuk (shape):", data.shape)     # (3, 2) â†’ 3 baris, 2 kolom
print("Jumlah elemen:", data.size)       # 6
print("Tipe data elemen:", data.dtype)   # int64
print("Total memori (byte):", data.nbytes)
```

Tipe Data

```python
import numpy as np

a = np.array([1, 2, 3], dtype=np.int32)
b = np.array([1, 2, 3], dtype=np.float64)
c = np.array([1, 2, 3], dtype=np.complex128)

print(a, a.dtype)   # [1 2 3] int32
print(b, b.dtype)   # [1. 2. 3.] float64
print(c, c.dtype)   # [1.+0.j 2.+0.j 3.+0.j] complex128
```

Membuat Array dengan Berbagai Bentuk

```python
print(np.array([1,2,3]))
print(np.zeros((2,3)))
print(np.ones(4))
print(np.arange(0,10,2))
print(np.linspace(0,1,5))
print(np.random.randint(0,10,(2,2)))
```

Membuat Matriks

```python
import numpy as np

M1 = np.array([[1, 2], [3, 4]])   # matriks manual
M2 = np.identity(3)               # matriks identitas 3x3
M3 = np.eye(3, k=1)               # diagonal di atas utama
M4 = np.diag([10, 20, 30])        # matriks diagonal

print(M1)
print(M2)
print(M3)
print(M4)
```
